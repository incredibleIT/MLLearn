### 一. 导入所需依赖

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import lineStyles
```

### 二. 定义需要的数据

```python
x = np.arange(0, 6, 0.1)
```

```python
y = np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
plt.show()
```

```python
y1 = np.cos(x)
print(y1)
```

```python
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y1")
plt.title("y1 = cos(x)")
plt.show()
```

### 三. 画出原函数与导函数的函数图像

```python
plt.plot(x, y, label="y=sin(x)")
plt.plot(x, y1, label="y'=cos(x)", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
