 ## **逻辑回归(logistic regression)**

logistic regression 是一种用于解决分类问题的统计方法, 尤其适用于二分类问题. 尽管名称中有回归, 但它主要用于分类任务.

logistic regression 将线性回归的输出映射到[0, 1]区间内来表示属于某个类别的概率.

常用的映射函数是sigmoid函数
$
\sigma(z) = \frac{1}{1 + e^{-z}}
$

\

逻辑回归的结果可以表示为
$$
P(y=1 \mid x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n)}}
$$

其中$\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n$为线性回归的输出结果.

$P(y=1 \mid x)$代表了输出为1类的概率, 根据逻辑回归的结果和阈值来确定最终的预测结果, 大于阈值的结果为1类, 小于阈值的结果为0类.

logistic regression Loss:

逻辑回归的损失函数一般是对数损失(log loss), 也称为二元交叉熵损失(Binary cross entropy loss). 用来衡量模型输出结果的概率分布与实际标签之间的差距, 逻辑回归的损失函数计算来源于最大似然函数(MLE).



似然函数 L(beta) 表示已知y的结果, 此时模型参数beta的概率分布, 简单来说就是在已知y这个结果的情况下, 此时参数为什么值时出现这个结果y的概率最大.



这里也就是求解Loss损失函数的最小值, 就可以达到最大似然函数的目的.也就是说当前参数是符合最大似然估计的最优参数.

对于逻辑回归的Loss求最小值, 可以对Loss求梯度, 之后通过梯度下降法来进行求解.



API

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(
    solver='sag',
    multi_class='multinomial',  # 多类别
    max_iter=1000,  # 迭代次数
    C=1.0, # 正则化参数
    class_weight='balanced',
    random_state=42,
    penalty='l1'
)
```
