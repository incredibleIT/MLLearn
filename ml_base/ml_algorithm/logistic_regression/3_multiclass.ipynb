## **多分类**

逻辑回归通常用于解决二分类问题, 但是可以通过一对多(one-vs-rest, ovr)以及Softmax回归(Multinomial Logistic Regression，多项逻辑回归)来扩展到多分类任务

#### 1. **OVR**

若有C个类别, 则训练C个二分类逻辑回归分类器, 每个分类器将C中的一个类别作为正类, 所有其他类别为负类, 进行预测时计算C个分类器的输出概率选择概率最高的类别作为输出

$$
P(y = c \mid x; \beta) = \frac{1}{1 + e^{-(\beta^T x)}}
$$

$$
Loss = -\frac{1}{n} \sum_{i=1}^n \left(
y_i \log P(y_i = c \mid x_i; \beta)
+ (1 - y_i) \log \left( 1 - P(y_i = c \mid x_i; \beta) \right)
\right)
$$

简单易于实现, 但是训练时间长

API

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression(multi_class="ovr")

from sklearn.multiclass import OneVsRestClassifier

one_vs_rest = OneVsRestClassifier(LogisticRegression())
```

#### 2. **Softmax Regression**

softmax函数(替换掉了sigmoid函数):

$$
P(y=c\mid x) = \frac{e^{\beta_c^\top x}}{\sum_{k=1}^C e^{\beta_k^\top x}}
$$

$$
\mathrm{Loss}=-\frac{1}{n}\sum_{i=1}^n\sum_{c=1}^C \mathbb{I}(y_i=c)\,\log P(y=c\mid x_i)
$$

只训练一个模型, 分类一致性好

API

```python
soft_max = LogisticRegression(multi_class="multinomial")
```
