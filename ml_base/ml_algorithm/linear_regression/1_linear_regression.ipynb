## **线性回归**

linear regression 是一种用于建模两个或多个变量之间线性关系的统计方法. 他通过拟合一条直线(或超平面)来描述自变量(输入特征)与因变量(输出目标)之间的关联, 并且可以用于预测, 分析变量间的影响关系.、

设因变量y与自变量(x1, x2, ..., xn)之间的关系可以由以下线性方程来描述:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n $$

其中, β0为截距项, β1, β2, ..., βn为回归系数, 表示每个自变量对因变量的影响程度.

通过估计这些参数来使模型预测值尽可能的接近真实值.


#### **1. 一元线性回归**

仅有一个自变量: $ y = \beta_0 + \beta_1 x $




#### **2. 多元线性回归**

包含多个自变量: $ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n $



#### **3. 简单的示例**

拟合一周内学习时间与考试分数之间的关系

```python
from sklearn.linear_model import LinearRegression   # 引入线性回归模型
```

```python
# 自变量，每周学习时长
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 因变量，数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]
```

```python
# 定义模型, 其中截距参数可以不设置, 通过fit_intercept=false来指定
model = LinearRegression()
```

```python
# 训练模型
model.fit(X, y)
print(f"截距项: {model.intercept_}")
print(f"回归系数: {model.coef_}")
```

```python
# 模型预测
x_new = [[11]]
y_pred = model.predict(x_new)
print(f"预测结果: {y_pred}")
```

```python
# 可视化
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_line = np.arange(0, 16, 0.1)
y_line = model.predict(x_line.reshape(-1, 1))
```

```python
plt.scatter(X, y)
plt.plot(x_line, y_line, c='r')
plt.scatter(x_new, y_pred, c='g')
plt.show()
```
