## **支持向量机**

支持向量机是一种二分类模型, 核心目标是找到一个间隔最大的超平面将不同类别的数据点分开.



间隔最大是指这个超平面与最近的数据点之间的距离最大, 这使得模型对训练数据集的泛化性能更好.

支持向量机有多种模型, 线性可分支持向量机, 线性支持向量机, 非线性支持向量机.

#### 线性可分支持向量机(硬间隔)

当训练集线性可分时, 此时可以通过最大化硬间隔来学习线性可分的向量机, 所谓硬间隔就是指超平面可以将数据集完全划分开. 距离超平面距离最近的几个样本点叫做支持向量, 支持向量决定了超平面的位置和方向, 支持向量不发生变化超平面就不会变



在样本空间中, 超平面可以表示为:

$$
\mathbf{w}^T \mathbf{x} + b = 0
$$
w 为法向量, b为超平面与原点的距离

对应的分类函数为:

$$
f(\mathbf{x}) = sign(\mathbf{w}^T \mathbf{x} + b)
$$

对于不同分类的样本到达超平面的距离为:
$$
r = \left| \frac{\mathbf{w}^T ( \mathbf{x} - \mathbf{x}' )}{\lVert \mathbf{w} \rVert} \right|
    = \frac{ \left| \mathbf{w}^T \mathbf{x} + b \right| }{ \lVert \mathbf{w} \rVert }
$$


记样本点xi的类别为yi, 该样本点的函数间隔为: $\hat{y}_i = y_i \left( \mathbf{w}^T \mathbf{x} + b \right)$, 若超平面可以将样本点正确分类则:





那么我们要找到一个超平面使得距离最近样本点的间隔最大, 即:



#### **线性支持向量机(软间隔)**



$$
\begin{aligned}
\min_{\mathbf{w},\, b,\, \xi} \quad & \frac{\lVert \mathbf{w} \rVert^2}{2} \;+\; C \sum_{i=1}^{n} \xi_i \\
\text{s.t.} \quad & y_i \left( \mathbf{w}^T \mathbf{x}_i + b \right) \ge 1 - \xi_i, \quad i = 1,2,\ldots,n
\end{aligned}
$$

为每个样本引入一个因子$\xi_i$, 条件变为了1-$\xi_i$, 即所有的点都在超平面内或者超平面外, 同时对误分类进行惩罚.

#### **非线性支持向量机**



通过核函数将数据从原始空间映射到高维特征空间, 使得数据在高维线性可分, 使用核技巧学习非线性支持向量机，等价于隐式地在高维特征空间中学习线性支持向量机。





核函数的选择也是支持向量机最大的变数，若核函数选择不合适，意味着将样本映射到了一个不合适的特征空间，很可能导致性能不佳
