3. **PCA**: 主成分分析

是一种常用的降维技术, 它将高维数据映射到低维空间, 并且保留原始数据的主要模式, 迭代的去旋转坐标轴让其落在数据方差最大的方向上这个方向叫做主数据,

继续在与pc1正交的子空间中寻找pc2, 迭代此操作, 实现降维

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# 生成数据
X = np.random.randn(1000, 3)
print(X.shape)
```

```python
# 使用PCA对数据进行降维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 转换前进行可视化
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c='g')
ax1.set_title('Before PCA')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')
# PCA后
ax2 = fig.add_subplot(122)
ax2.scatter(X[:, 0], X[:, 1], c='g')
ax2.set_title('After PCA')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
plt.show()
```

为了使结果更明显, 可以手动构建数据集

```python
# 标准正态分布
pc1 = np.random.normal(0, 1, 1000)
# 特殊方差正态分布
pc2 = np.random.normal(0, 0.5, 1000)
# 方差更小的分布
noise = np.random.normal(0, 0.01, 1000)
```

```python
# 构造feature矩阵
X = np.vstack((pc1 + pc2, pc1 - pc2, pc2 + noise)).T
print(X.shape)
```

```python
# 使用PCA对数据进行降维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 转换前进行可视化
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c='g')
ax1.set_title('Before PCA')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')
# PCA后
ax2 = fig.add_subplot(122)
ax2.scatter(X[:, 0], X[:, 1], c='g')
ax2.set_title('After PCA')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
plt.show()
```
