B. **斯皮尔曼相关系数**

斯皮尔曼相关系数: 是用来衡量两个变量之间的**单调关系monotonic relationship**强度的统计指标。当一个变量增加时, 另一个变量是否总是增加或减少但是并不要求线性关系, 适用于非线性或不符合正态分布的数据。

范围为: [-1, 1], 值越接近1, 表示正相关越强, 值越接近0, 表示越弱, 值越接近-1, 表示负相关越强。

$$
\rho = 1 - \frac{6 \sum_{i=1}^{n} d_i^2}{n (n^2 - 1)}
$$


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 创建数据
# 固定随机种子，确保结果可复现
np.random.seed(42)

# 构造基础变量 x
n = 20
x = np.arange(1, n + 1)

# 构造不同类型的 y
df = pd.DataFrame({
    "x": x,
    "y_perfect_pos": x,                      # 完全正单调
    "y_perfect_neg": -x,                     # 完全负单调
    "y_monotonic_nl": np.log(x + 1) * 10,    # 非线性但单调
    "y_noise": np.random.normal(0, 1, n),    # 噪声
    "y_ties": [1]*5 + [2]*5 + [3]*5 + [4]*5, # 含平局
    "y_partial": x + np.random.normal(0, 3, n) # 部分单调
})
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```
