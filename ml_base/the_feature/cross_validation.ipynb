## **交叉验证**

是一种评估模型泛化能力的方法, 通过将数据集划分为多个子集, 反复的进行训练和验证, 以避免单次数据划分带来的随机性误差

#### **1. 简单交叉验证**

将数据划分为训练集和验证集, 结果受单次划分的影响较大可能高估, 低估模型的性能

#### **2. k折交叉验证**

将数据集均匀的划分为k折, 每次用(k - 1)折进行训练, 用剩余的1折进行验证, 循环k次, 最后取平均性能, 对数据运用更加充分, 结果更稳定

数据集 D 被划分为 10 份：

D = [ D₁ | D₂ | D₃ | D₄ | D₅ | D₆ | D₇ | D₈ | D₉ | D₁₀ ]

每次取一个子集为验证集，其余为训练集：

| 轮次 | 训练集                              | 验证集 | 结果       |
|------|--------------------------------------|--------|------------|
| 1    | D₁~D₉                               | D₁₀    | 测试结果₁ |
| 2    | D₁~D₈, D₁₀                         | D₉     | 测试结果₂ |
| 3    | D₁~D₇, D₉, D₁₀                     | D₈     | 测试结果₃ |
| ...  | ...                                  | ...    | ...        |
| 10   | D₂~D₁₀                              | D₁     | 测试结果₁₀ |

最终平均：
L_cv = (测试结果₁ + 测试结果₂ + … + 测试结果₁₀) / 10


#### **4. 留一交叉验证**

每次仅留一个样本作为验证集, 其余样本全部用作训练, 重复动作直到所有样本都被验证一次结束, 适用于小数据集, 计算成本高
