### 一. 基本术语
1. 数据集(Data set): 训练集(用于训练模型的数据), 测试集(用于评估模型效果的数据), 验证集(用于调节超参数的数据)

2. 样本(Sample): 数据集中的一条记录是关于一个事件或者对象的描述, 称之为一个样本
3. 特征(Feature): 数据集中一列反映事件或对象在某方面的表现或性质的事项, 为特征或者属性
4. 特征向量(Feature Vector): 将样本中的所有特征表示为向量的形式, 输入到模型中, 这个向量形式的数据就是特征向量
5. 标签(label): 监督学习中每个样本的结果信息, 也称之为目标值(target)
6. 模型(Model): 模型是一个机器学习算法与训练后的参数的集合, 用于对新的数据预测或分类
7. 参数(Parameter): 模型通过训练学习到的值, 例如线性回归中的权重值和偏置项
8. 超参数(Hyper Parameter): 用户设置的参数, 不能通过训练自动学习, 例如:学习率, 迭代次数, 模型结构, 正则系数等

### 二. 基本理论
##### 机器学习方法 = 模型 + 策略 + 算法
1. 模型: 模型是机器学习算法的输入, 模型描述了如何处理数据, 模型是学习算法的输入, 模型描述了如何处理数据

2. 策略: 选取最优模型的评价准则
3. 算法: 选取最优模型的具体方法
##### 机器学习算法分类: 有监督学习, 无监督学习, 强化学习
1. 有监督学习: 训练数据中**包含标签信息**, 模型需要根据训练数据中的标签信息进行训练, 主要分为线性分类与回归

2. 无监督学习: 训练数据中**不包含标签信息**, 模型需要根据训练数据进行训练(无监督分组, 聚类)
3. 强化学习: 通过环境与智能体进行交互, 智能体根据环境信息进行选择行为


### 三. 建模流程
1. 收集数据: 收集用于训练和测试的数据集, 确保代表了实际问题的不同方面

2. 数据清洗: 对数据进行清洗, 去除掉一些脏数据和不可用数据
3. 特征工程: 对数据进行转换与格式化确保适合用于模拟训练
4. 选择算法: 选择适合任务类型与数据特征的机器学习算法
5. 模型训练: 使用**训练数据集**来训练模型, 让模型从数据中学习规律与模式
6. 模型评估: 使用**测试数据集**来评估模型性能确定是否达到了目标
7. 模型优化: 根据评估结果对模型进行优化, 尝试不同的算法或参数, 直到获得更好的结果
8. 模型部署: 将模型部署到实际应用中, 确保模型在部署时能达到预期的效果

### 四. 特征工程: 转换, 构造, 处理 原始数据为更好表示问题的特征形式
1. 特征选择: 从原始特征中挑选出与目标变量关系最密切的特征, 剔除冗余无关的噪声, 不会创建新特征也不会改变数据结构
- 过滤法: 基于统计测试(低方差过滤法, 卡方检验, 相关系数)

- 包裹法: 例如递归特征消除, 来评估特征的重要性, 并且根据模型的表现进行特征的选择
- 嵌入法: 使用模型本身的特征选择机制(如决策树特征的重要性,L1正则化的特征选择)来选择最重要的特征
2. 特征转换: 对数据进行数学或者统计处理, 使其更符合模型的输入要求
- 归一化: 将特征缩放到特定的范围内, 通常是[0, 1]之间, 适用于对尺度敏感的类型

- 标准化: 通过减去平均值除去标准差, 使特征分布具有均值0和标准差1的情况
- 对数变换: 对于有偏态的分布(例如收入), 通过对数变换可以将其转化为更接近正态分布的情况
- 类别变量编码: 将类别变量转换为数值变量, 例如将类别变量转换为独热编码
    - 独热编码(one-hot-encoding): 将类别型变量转换为二进制列, 常用于**无序类别特征**

    - 标签编码(label-encoding): 将类别型变量映射为整数, 常用于**有序类别特征**
    - 目标编码(target-encoding): 将类别型变量映射为目标变量的均值
    - 频率编码(frequency-encoding): 将类别型变量映射为类别在数据集中出现的频率
3. 特征构造: 基于现有的特征创造出新的更具代表性的特征, 组合转换聚合现有特征形成能够更好反映数据规律的特征
    - 交互特性: 将两个特征组合起来形成新的特征

    - 统计特征: 从原始特征中提取统计值, 例如平均值, 最大值, 最小值, 标准差等
    - 日期和时间特征: 从日期时间数据中提取如星期几, 季度, 月份等
4. 特征降维: 当数据集特征数量非常庞大时, 通过降维的方法可以保持数据本质的情况下减少特征的数量
    - 主成分分析(PCA): 通过线性变换将原始特征映射到一个新的空间, 使得新的特征(主成分)尽可能的保留数据的方差

    - 线性判别分析(LDA): 一种监督学习的降维方法
