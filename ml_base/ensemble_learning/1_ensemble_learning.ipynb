## **集成学习**

集成学习通过某种策略通过组合多个个体学习器的预测结果来提升整体的预测能力.只包含同种类型的个体学习器的集成叫做同质集成, 同质集成中的个体学习器也叫做基学习器, 相应的学习算法称为基学习算法。包含不同类型的个体学习器的集成称为异质集成.

集成学习有三大经典的方法: Boosting(提升方法), Bagging(自助聚合), Stacking(堆叠)

Boosting: 顺序的训练模型, 每一个模型关注前一个模型的错误, 通过加权调整来优化整体的预测

Bagging: 从原始数据集中通过有放回的对样本采样生成多个子数据集，分别训练多个独立模型，最后通过投票（分类）或平均（回归）得到结果

Stacking: 训练多个不同类型的个体学习器，之后使用一个元模型综合多个个体学习器的预测

#### **AdaBoost**

AdaBoost通常使用单层决策树作为基学习器，单层决策树也被称为决策树桩（Decision Stump）

AdaBoost的做法是提高被前一轮弱分类器分错的样本的权重, 降低分正确的样本权重, 这样权重较大的样本会被下一个模型更加重视, 因为大权重对loss function影响更大优先被拟合, 同时采用加权多数表决的方法，加大分类误差率小的弱分类器的权重，减小分类误差率大的弱分类器的权重。



#### **随机森林**

随机森林是Bagging的一个变体，在以决策树为基学习器构建Bagging集成的基础上，进一步在决策树训练过程中引入了随机属性选择。

在传统的决策树中, 每一个节点是在当前特征集合中选择的是最优特征, 而在随机森林中基决策树每个节点先从该节点的特征集合中选取k个特征子集之后在这个子集中选取最优特征进行划分

参数k控制着随机性的引入程度，若k=d，则基决策树的生成与传统决策树相同；若k=1，则随机选择一个属性用于划分。一般推荐k=log2d。
