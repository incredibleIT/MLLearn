## 概率论


概率分布
1. 均匀分布: 1/(b - a)
2. 正态分布(高斯分布)

```python
import numpy as np
```

```python
# 随机生成[0, 1]均匀分布数据
a = np.random.rand(5)
print(a)
```

```python
# 随机生成均匀分布整数
b = np.random.randint(0, 100, size=[2, 3])
print(b)
```

```python
# 随机生成标准正态分布
c = np.random.randn(2, 5)
print(c)
```

```python
# 随机生成普通正态分布数据
d = np.random.normal(loc=5, scale=1, size=[2, 5])
print(d)
```

### 贝叶斯定理: 从结果反推原因的概率
全概率公式: P(B) = P(B|A1) * P(A1) + P(B|A2) * P(A2) + P(B|A3) * P(A3) + ... + P(B|An) * P(An)
贝叶斯: P(A|B) = P(B|A) * P(A) / P(B)

### 似然函数
概率是在一些已知某些观测值的情况下预测观测上的结果, 似然性是在某些观测得到的结果时对有关事物性质的参数进行估值
L(sita|X) = P(X|sita)

### 极大似然估计
似然函数的极大值就是估计的参数
